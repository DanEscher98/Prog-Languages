```python
#The relationship between gravitational force 
#and distance between two bodies

import matplotlib.pyplot as plt

# Draw the graph
def draw_graph(x, y):
    plt.plot(x, y, marker='o')
    plt.xlabel('Distance in meters')
    plt.ylabel('Gravitational force in newtons')
    plt.title('Gravitational force and distance')
    plt.show()
    
def generate_F_r(m1, m2):
    r = range(100, 1001, 50)
    F = []
    G = 6.674*(10**-11)
    
    # Calculate force and add it to the list, F
    for dist in r:
        force = G*(m1*m2)/(dist**2)
        F.append(force)
    draw_graph(r, F)
    
if __name__=='__main__':
    m1 = float(input('Masa de objeto 1: '))
    m2 = float(input('Masa de objeto 2: '))
    generate_F_r(m1, m2)
```

```python
# Draw the trajectory of a body in projectile motion

from matplotlib import pyplot as plt
import math

def draw_graph(x, y, name):
    plt.plot(x, y)
    plt.xlabel('x-coordinate')
    plt.ylabel('y-coordinate')
    plt.title('Projectile motion')
    plt.savefig('Output/'+name+'.png')
    plt.legend([name])

def frange(start, final, interval):
    numbers = []
    while start<final:
        numbers.append(start)
        start += interval
    return numbers

def draw_trajectory(u, theta):
    name = 'init-v = '+str(u)+'\nangle = '+str(theta)
    theta = math.radians(theta)
    g = 9.8
    t_flight = 2*u*math.sin(theta)/g
    intervals = frange(0, t_flight, 0.001)
    x = []; y = []
    for t in intervals:
        x.append(u*math.cos(theta)*t)
        y.append(u*math.sin(theta)*t - 0.5*g*t*t)
    draw_graph(x, y, name)
    
if __name__=='__main__':
    try:
        u = float(input('Enter the initial velocity (m/s): '))
        theta = float(input('Enter the angle of projection (degrees): '))
    except ValueError:
        print('You entered an invalid input')
    else: 
        draw_trajectory(u, theta)
        plt.show()
```

```python
#Calculating the mean
from collections import Counter

def calc_mean(numbers):
    s = sum(numbers)
    n = len(numbers)
    return s/n

def calc_median(numbers):
    N = len(numbers)
    numbers.sort()
    if N%2 == 0:
        m1=int(N/2)-1; m2=int(N/2)
        median = (numbers[m1]+numbers[m2])/2
    else:
        m = int((N+1)/2)-1
        median = numbers[m]
    return median

def calc_mode(numbers):
    c = Counter(numbers)
    num_freq = c.most_common()
    max_count = num_freq[0][1]
    modes = []
    for num in num_freq:
        if num[1] == max_count:
            modes.append(num[0])
    return modes

def freq_table(numbers):
    table = Counter(numbers)
    num_freq = table.most_common()
    #num_freq.sort()
    print('Number\tFrequency')
    for number in num_freq:
        print('{0}\t{1}'.format(number[0], number[1]))

if __name__=='__main__':
    data = [100, 234, 322, 102, 100, 320, 432, 210, 150, 102, 432, 432]
    print('Mean is {}'.format(calc_mean(data)))
    print('Median is: {}'.format(calc_median(data)))
    print('Modes are: {}'.format(calc_mode(data)))
    freq_table(data)
    
```
